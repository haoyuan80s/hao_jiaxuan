```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2 as cv
import glob
```

```python
images = glob.glob('data/train_images/*')
images = np.random.permutation(images)
```

```python
def vis(fn):
    im = cv.imread(fn)
    print(im.shape)
    plt.imshow(im)

fig = plt.figure(figsize=(15, 15))
for i, im in enumerate(images[:6]):
    fig.add_subplot(2,3,i+1)
    plt.axis('off')
    vis(im)
plt.show()
```

```python
num_chars = []
with open('data/train.csv') as f:
    first = True
    for line in f.readlines():
        if first: 
            first = False
            continue
        
        info = line.split(",")
        num_chars.append(len(info[1].split()) / 5)
    
sns.distplot(num_chars)
plt.title('number of characters in an image')
plt.show()
        
print(np.mean(num_chars))
```
